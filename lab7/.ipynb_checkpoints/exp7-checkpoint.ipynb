实验七 Linux实验教学系统开发
=======

# 1. 实验类型

设计型，8学时，必做实验


# 2.  实验目的

通过Shell编程解决一个实际问题，理解Shell程序设计在系统管理方面的应用，分析掌握软件项目管理，Shell程序设计、编写的基本技术。

# 3. 题目描述

本实验以使用Shell程序设计解决本课程的实验教学系统管理方面的任务为内容，通过分析、讨论、设计、调试、报告过程，最后完成项目计划、设计文档、程序代码以及报告材料。

# 4. 实验要求

* 基本层次要求：参与分析、讨论、设计、调试、报告过程，完成要求的项目计划、设计文档、程序代码和报告材料，要求材料符合规范，内容充实，代码运行正确，解决实际问题；

* 扩展层次要求：针对设计结果的不足之处，提出解决方案，要求设计并测试解决方案。

# 5. 相关知识

## 5.1 如何进行项目计划？

软件项目计划是一个软件项目能够得以完成的重要保障，不管对科学研究还是对生产实践，项目计划都有至关重要的作用，学习如何进行项目的计划具有非常重要的意义。

项目计划一般包含项目题目、目标、任务、问题、资源需求、任务计划、时间－任务图（甘特图）等。项目题目是项目的一句话的描述；项目目标是项目方向的概要描述，一般只有一到两个；任务是围绕项目目标的具体的细化的项目内容，一般有很多；问题是与项目开发相关的难题，条件限制等的讨论；资源需求是进行开发所需要的硬件及软件资源的保障；任务计划是以列表的方式列出需要解决的问题、参与的人员、目标等；甘特图是一个线条图，横轴表示时间,纵轴表示活动（项目），线条表示在整个期间上计划和实际的活动完成情况，它直观地表明任务计划在什么时候进行，及实际进展与计划要求的对比。比如一个图书出版的甘特图可能如下图7-1：

![图书出版的甘特图](images/7-1.png)
<center>图7-1 图书出版的甘特图</center>

## 5.2 如何进行讨论？

讨论是实现项目的重要保证，讨论前需要做好计划，包括时间、地点、参加人、讨论内容等，需要由讨论组织者进行讨论计划的书写并提前发到参与讨论人员手上，讨论时需要有记录者、组织者，讨论结束后需要指定人员完成讨论总结，发到每个参与者手中。讨论总结应包括对讨论计划的实施或者安排结果，需要表现出讨论的意义。

## 5.3 如何完成材料？

报告材料主要指面对很多听众做口头报告时需要准备的文档，通常用投影机做报告需要准备power point文档，其中包括对题目的分析和程序设计，以及重要问题的提出和解决方法。

设计文档是对程序设计过程的总结，其中包括流程图、模块划分、实现、重要问题的讨论和实现等。国家标准对软件开发规范有详细的定义，其中包含对详细设计说明书和项目开发总结有一个标准格式的定义，本次实验综合了国家标准的相关文档，具体参考报告模板中的内容report-template.ipynb。

# 7. 实验指导

“Linux实验教学系统”是“Linux系统及程序设计”课程组开发的一个旨在提高该课程实验效果的应用系统（源自：西南科技大学实验技术研究项目“面向MOOC的Linux课程实验教学技术研究”，项目编号：15syjs-22）。系统部署在网络与安全实验室，可以通过互联网访问，方便学生进行自主学习，如图7-3：

![系统部署](images/7-3.png)

系统采用了容器技术和交互式文档技术，为学生提供一个独立的、交互式的Linux实验环境，并通过Bash脚本实现了实验环境的管理，包括启动、停止、提交等，每个实验的实验数据和实验管理脚本构成该实验的实验包，通过项目开发网站进行版本管理，如图7-4：

# 8、实验任务

根据“Linux实验教学系统”待开发功能列表（可以在Gitlab中linux2016/lab7.git项目空间中的issues功能中查看）：

1）学生端

* 下载提交数据

* 查看成绩

* 备份提交数据

* 检查提交版本是否为最新版本

* 集中显示查看实验中所有任务并检查实验报告是否完成任务

* 单个实验成绩或综合实验成绩不合格预警

2）教师端

* 成绩统计

* 管理功能集成

* 分组实验管理功能设计与开发

* 单个实验成绩或综合实验成绩不合格预警 

选择列表中的一个待开发功能（学生端或教师端中的任意一个），讨论具体开发需求，使用Gitlab工具进行团队开发（3人一组），实现功能，完成开发总结报告一份、代码一份，并进行口头报告。口头报告的内容应包括：任务分工、时间计划、代码运行、程序结构、遇到的问题和解决等。
